# Evaluación de la calidad de clusterizado mediante métricas de distancia

Objetivo: El/La estudiante sea capaz de evaluar y comparar la calidad de distintas configuraciones de clusterizado aplicando metricas de distancia

```python
#Librerias
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, pairwise_distances
import matplotlib.pyplot as plt
```

El presente ejemplo busca implementar métricas de calidad de clusterizado a través de uso del algoritmo KMEANS sobre un conjunto de datos clásico llamado Iris

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
silhouette_list=[]
davies_list=[]
calinski_list=[]

for n_clusters in range(2,11):
    K_means = KMeans(n_clusters,random_state=42)
    labels = K_means.fit_predict(iris.data)
    centroides = K_means.cluster_centers_

    #Indice de métricas de distancia
    silhouette = silhouette_score(iris.data,labels)
    silhouette_list.append(silhouette)

    davies_bouldin = davies_bouldin_score(iris.data,labels)
    davies_list.append(davies_bouldin)

    calinski_harabasz = calinski_harabasz_score(iris.data,labels)
    calinski_list.append(calinski_harabasz)

    print('numero de clusters: ', n_clusters)
    print('Indice silhouette: ', silhouette)
    print('Indice davies: ', davies_bouldin)
    print('Indice calinski: ', calinski_harabasz)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(range(2,11), silhouette_list, marker='o')
plt.title("Indice Silhouette")
plt.xlabel("Numero de clusters")
plt.ylabel("Numero de indices")

plt.subplot(1,3,2)
plt.plot(range(2,11), davies_list, marker='o')
plt.title("Indice Davies")
plt.xlabel("Numero de clusters")
plt.ylabel("Numero de indices")

plt.subplot(1,3,3)
plt.plot(range(2,11), calinski_list, marker='o')
plt.title("Indice Calinski")
plt.xlabel("Numero de clusters")
plt.ylabel("Numero de indices")

plt.tight_layout()
plt.show()
```
