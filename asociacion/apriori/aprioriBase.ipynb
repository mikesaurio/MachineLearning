```python
transacciones = df['Items'].str.split(',').tolist()
print(transacciones[:5])
```

```python
conjuntos, reglas = apriori(transacciones, min_support=0.01, min_confidence=0.05)
```

```python
for i, regla in enumerate(reglas):
    print(f"Regla {i+1}: {regla}")
```

```python
reglas = list(filter(lambda regla: len(regla.rhs) == 1, reglas))
```
