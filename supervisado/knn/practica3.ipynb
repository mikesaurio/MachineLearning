# Practica 3

Una empresa de comercio electrónico está interesada en analizar el comportamiento de sus 
clientes durante campañas de marketing digital, con el objetivo de predecir si un usuario 
realizará una compra (1) o no (0). 

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, cross_val_score, train_test_split
```

1) Lea el archivo csv “marketing_digital.csv” y cárguelo en un Dataframe llamado data, imprima 
los primeros 5 registros para revisar que se haya cargado de manera correcta. 

```python
data = pd.read_csv("data/marketing_digital.csv")
data.head()
```

2) Calcule la proporción de cada clase de la variable objetivo 

```python
data.describe()
```

```python
data["Compra"].value_counts(normalize=True)
```

3) Defina variables predictoras y la variable objetivo. 

```python
X=data.iloc[:,0:6]
X.head()
```

```python
Y=data.iloc[:,6]
Y.head()
```

4) Escale las variables predictoras para estandarizarlas 

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

5) Construya el modelo KNN (use 5 vecinos), SVC y el de árboles de clasificación 

```python
knn=KNeighborsClassifier(n_neighbors=5, metric="euclidean")
```

```python
svc=SVC(kernel='rbf', C=1.0, gamma="scale", random_state=42)
```

```python
tree=DecisionTreeClassifier(max_depth=3)
```

6) Aplique validación cruzada con 5 folds 

```python
cv=StratifiedKFold(n_splits=5,shuffle=True, random_state=42)
```

7) Obtenga accuracy para los 3 modelos 

```python
accuracy_score_knn=cross_val_score(knn,X_scaled, Y,cv=cv,scoring="accuracy")
accuracy_score_svm=cross_val_score(svc,X_scaled, Y,cv=cv,scoring="accuracy")
accuracy_score_tree = cross_val_score(tree, X, Y, cv=cv,scoring="accuracy") 
```

```python
print("KNN")
print("Accuracy para cada bloque",accuracy_score_knn)
print("Accuracy promedio",accuracy_score_knn.mean())
print("Accuracy desviación",accuracy_score_knn.std())
print("-"*10)
print("SVM")
print("Accuracy para cada bloque",accuracy_score_svm)
print("Accuracy promedio",accuracy_score_svm.mean())
print("Accuracy desviación",accuracy_score_svm.std())
print("tree")
print("-"*10)
print("Accuracy para cada bloque",accuracy_score_tree)
print("Accuracy promedio",accuracy_score_tree.mean())
print("Accuracy desviación",accuracy_score_tree.std())
print("-"*10)
```

8) Obtenga f1 para los 3 modelos 

```python
f1_score_knn=cross_val_score(knn,X_scaled, Y,cv=cv,scoring="f1")
f1_score_svm=cross_val_score(svc,X_scaled, Y,cv=cv,scoring="f1")
f1_score_tree = cross_val_score(tree, X, Y, cv=cv,scoring="f1") 
```

```python
print("KNN")
print("Accuracy para cada bloque",f1_score_knn)
print("Accuracy promedio",f1_score_knn.mean())
print("Accuracy desviación",f1_score_knn.std())
print("-"*10)
print("SVM")
print("Accuracy para cada bloque",f1_score_svm)
print("Accuracy promedio",f1_score_svm.mean())
print("Accuracy desviación",f1_score_svm.std())
print("tree")
print("-"*10)
print("Accuracy para cada bloque",f1_score_tree)
print("Accuracy promedio",f1_score_tree.mean())
print("Accuracy desviación",f1_score_tree.std())
print("-"*10)
```

9) Repita el proceso, pero ahora utilice 9 vecinos.
