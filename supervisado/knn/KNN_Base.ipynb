```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
data = pd.read_csv("data/diabetes.csv")
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data.describe()
```

```python
data["Diabetes"].value_counts(normalize=True)
```

```python
X=data.iloc[:,0:8]
X.head()
```

```python
Y=data.iloc[:,8]
Y.head()
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
knn=KNeighborsClassifier(n_neighbors=9, metric="euclidean")
```

```python
cv=StratifiedKFold(n_splits=5,shuffle=True, random_state=42)
```

```python
accuracy_score_knn=cross_val_score(knn,X_scaled, Y,cv=cv,scoring="accuracy")
```

```python
print("Accuracy para cada bloque",accuracy_score_knn)
print("Accuracy promedio",accuracy_score_knn.mean())
print("Accuracy desviación",accuracy_score_knn.std())
```

```python
accuracy_score_knn=cross_val_score(knn,X_scaled, Y,cv=cv,scoring="f1")
```

```python
print("F1 para cada bloque",accuracy_score_knn)
print("F1 promedio",accuracy_score_knn.mean())
print("F1 desviación",accuracy_score_knn.std())
```

# SVM

```python
svm=SVC(kernel='rbf', C=1.0, gamma="scale", random_state=42)
```

```python
accuracy_score_svm=cross_val_score(svm,X_scaled, Y,cv=cv,scoring="accuracy")
```

```python
print("Accuracy para cada bloque",accuracy_score_svm)
print("Accuracy promedio",accuracy_score_svm.mean())
print("Accuracy desviación",accuracy_score_svm.std())
```

```python
accuracy_score_svm=cross_val_score(svm,X_scaled, Y,cv=cv,scoring="f1")
```

```python
print("F1 para cada bloque",accuracy_score_svm)
print("F1 promedio",accuracy_score_svm.mean())
print("F1 desviación",accuracy_score_svm.std())
```
